```python
def barra_carga_mision(texto, duracion=10):
    total = 50 # Cantidad total de bloques que tendrá la barra
    tiempo_por_paso = duracion / total # Tiempo que tarda cada actualización

    print(texto)
    print()

    # Bucle que construye la barra progresivamente
    for i in range(total + 1):

        # Construye la parte llena y la parte vacía de la barra
        barra = "█" * i + "░" * (total - i)

        # Calcula el porcentaje de progreso
        porcentaje = int((i / total) * 100)

        # \r devuelve el cursor al inicio de la línea para sobrescribirla
        sys.stdout.write(f"\r{barra} {porcentaje}%")

        # Fuerza a que se imprima inmediatamente en pantalla
        sys.stdout.flush()

        # Pausa el programa para simular progreso
        time.sleep(tiempo_por_paso)
    print("\n")
```

```python
def adivinanza(lv1, vidas, pistas_restantes,adf, ad2, nom):

    # Si el nivel es fácil
    if lv1 == 1:
        n = "1f"

        encabezado()
        frase(vidas,nom , pistas_restantes,n)

        # Mientras el jugador tenga vidas
        while vidas > 0:
            reto = random.choice(adf)
            intentos = 0
            f = True

            while f: # Bucle hasta acertar o fallar

                encabezado2(vidas, pistas_restantes)
                print("La adivinanza es:\n")
                print(reto["pregunta"])
                respuesta = input("\nRespuesta: ").upper()

                if not respuesta.isalpha(): # Si la entrada no son solo letras
                    print("⚠️ Solo se permiten letras.")
                    time.sleep(5)
                    clear_output()
                    continue

                if respuesta == "PISTA": # Si pide pista
                    if pistas_restantes > 0:
                        print("\nPista:", reto["pistas"][2 - pistas_restantes])
                        pistas_restantes -= 1
                        time.sleep(5)
                        clear_output()
                        continue
                    else:
                        print(f"\n⚠️ {nom} ya no te quedan pistas.")
                        time.sleep(5)
                        clear_output()
                        continue

                if respuesta == reto["respuesta"]: # Si responde correctamente, gana la prueba
                    print(f"\n¡Correcto! Superaste la adivinanza, bien hecho {nom}.")
                    time.sleep(5)
                    clear_output()
                    return vidas, pistas_restantes
                else: # Si falla, pierde vida
                    vidas -= 1
                    print("\nIncorrecto. Pierdes una vida💔.")
                    print("Vidas restantes❤️️:", vidas)
                    f = False
                    time.sleep(5)
                    clear_output()

        return vidas, pistas_restantes

    else: # Si el nivel es difícil
        n = "1d"

        encabezado()
        frase(vidas,nom , pistas_restantes,n)

        # Mientras tenga vidas
        while vidas > 0:
            reto2 = random.choice(ad2)
            intentos = 0
            f = True

            while f: # Bucle hasta acertar o fallar

                encabezado2(vidas, pistas_restantes)
                print("La adivinanza es:\n")
                print(reto2["pregunta"])
                respuesta = input("\nRespuesta:").upper()

                if not respuesta.isalpha(): # Si la entrada no son solo letras
                    print("⚠️ Solo se permiten letras.")
                    time.sleep(5)
                    clear_output()
                    continue

                if respuesta == "PISTA": # Si pide pista
                    if pistas_restantes > 0:
                        print("\nPista:", reto2["pistas"][2 - pistas_restantes])
                        pistas_restantes -= 1
                        time.sleep(5)
                        clear_output()
                        continue
                    else:
                        print(f"\n⚠️ {nom} ya no te quedan pistas.")
                        time.sleep(5)
                        clear_output()
                        continue

                if respuesta == reto2["respuesta"]: # Si responde correctamente, gana
                    print(f"\n¡Correcto! Superaste la adivinanza, bien hecho {nom}.")
                    time.sleep(5)
                    clear_output()
                    return vidas, pistas_restantes
                else: # Si falla, pierde vida
                    vidas -= 1
                    print("\nIncorrecto. Pierdes una vida💔.")
                    print("\nVidas restantes❤️️:", vidas)
                    f = False
                    time.sleep(5)
                    clear_output()

        return vidas, pistas_restantes, n
```

```python
def turno_maquina(tablero, lv1):

    disponibles = [(i, j) for i in range(3) for j in range(3) if tablero[i][j] == " "] # Evalua que casillas estan vacias

    if lv1 == 1: # Triqui en nivel dificil
        fila, col = random.choice(disponibles) # tomar una ubicaion aleatorio de los casillas vacias
        tablero[fila][col] = "O"

    else: # Triqui en nivel dificil

        # Intentar ganar y generar un movimiento que va a ganar
        for i, j in disponibles:
            tablero[i][j] = "O"
            if verificar_ganador(tablero) == "O":
                return
            tablero[i][j] = " "

        # Bloquear jugador
        for i, j in disponibles:
            tablero[i][j] = "X"
            if verificar_ganador(tablero) == "X":
                tablero[i][j] = "O"
                return
            tablero[i][j] = " "

        fila, col = random.choice(disponibles)
        tablero[fila][col] = "O"
```

```python
def turno_jugador(tablero, posiciones,vidas, pistas_restantes):

    # Bucle hasta que el jugador haga una jugada válida
    while True:
        try:
            opcion = input("Selecciona una posición (1-9): ")

            if opcion.upper() == "GUIA":
                clear_output()
                mostrar_guia2()
                time.sleep(5)
                clear_output()

                encabezado2(vidas, pistas_restantes)
                mostrar_tablero(tablero)
                continue

            opcion = int(opcion)
            # Si el número no está dentro de las posiciones válidas
            if opcion not in posiciones:
                print("⚠️ Número inválido.")
                continue

            fila, columna = posiciones[opcion]

            # Si la casilla está libre, coloca la X y termina el turno
            if tablero[fila][columna] == " ":
                tablero[fila][columna] = "X"
                break
            else: #Si la casiila esta ocupada
                print("⚠️ Esa casilla ya está ocupada.")

        # Si el usuario escribe algo que no es número
        except ValueError:
            print("⚠️ Solo números del 1 al 9.")
```

```python
def verificar_ganador(tablero):
 # Verifica si hay ganador por filas
    for fila in tablero:
        # Si las tres posiciones de la fila son iguales y no están vacías
        if fila[0] == fila[1] == fila[2] != " ":
            # Retorna el símbolo ganador ("X" o "O")
            return fila[0]

    # Verifica si hay ganador por columnas
    for col in range(3):
        # Compara las tres posiciones de cada columna
        if tablero[0][col] == tablero[1][col] == tablero[2][col] != " ":
            # Retorna el símbolo ganador
            return tablero[0][col]

    # Verifica la diagonal principal (de arriba izquierda a abajo derecha)
    if tablero[0][0] == tablero[1][1] == tablero[2][2] != " ":
        # Retorna el símbolo ganador
        return tablero[0][0]

    # Verifica la diagonal secundaria (de arriba derecha a abajo izquierda)
    if tablero[0][2] == tablero[1][1] == tablero[2][0] != " ":
        # Retorna el símbolo ganador
        return tablero[0][2]

    return None
```

```python
def tablero_lleno(tablero):
    # Verifica si el tablero esta lleno y no hay ganardor
    for fila in tablero:
        if " " in fila:
            return False
    return True
```

```python
def triqui(lv1, vidas, pistas_restantes, nom):

    # Obtiene las posiciones válidas del tablero
    posiciones = coor()
    if lv1 == 1:
        n = "3f"
    else:
        n = "3d"

    # Muestra la guía inicial del juego con vidas, pistas y nombre
    mostrar_guia(vidas, pistas_restantes,nom, n)
    time.sleep(5)

    # Mientras el jugador tenga vidas se puede jugar
    while vidas >0:

        # Crea un tablero nuevo para cada partida
        tablero = crear_tablero()

        # Inicia partida
        while True:

            clear_output()
            encabezado2(vidas, pistas_restantes)
            mostrar_tablero(tablero)

            # Turno del jugador (X)
            turno_jugador(tablero, posiciones,vidas, pistas_restantes)

            # Verifica si el jugador ganó
            if verificar_ganador(tablero) == "X":
                clear_output()
                encabezado2(vidas, pistas_restantes)
                mostrar_tablero(tablero)
                print("🎉¡Superaste la prueba del triqui!")
                o = True
                time.sleep(5)
                clear_output()
                return vidas, pistas_restantes  # Retorna las vidas y pistas restantes al ganar

             # Verifica si el tablero está lleno (empate y reinicia la partida)
            if tablero_lleno(tablero):
                print("🤝 Empate, se reinicia la partida.")
                time.sleep(2)
                break

            # Turno de la máquina (O)
            turno_maquina(tablero, lv1)

            # Guarda el ganador (si lo hay)
            ganador = verificar_ganador(tablero)

            # Si la máquina gana
            if ganador == "O":
                vidas -= 1 # Quita una vida
                clear_output()
                encabezado2(vidas, pistas_restantes)
                mostrar_tablero(tablero)
                print("💀 La máquina ganó.")
                print("❌ Pierdes una vida.")
                time.sleep(2)
                break # Reinicia la partida

            # Verifica empate después del turno de la máquina
            if tablero_lleno(tablero):
                print("🤝 Empate... se reinicia la partida.")
                time.sleep(2)
                break   # Reinicia la partida

    # Cuando se acaban las vidas, retorna los valores finales
    return vidas, pistas_restantes
```

```python
def decision(vidas, pistas_restantes, lv1, nom):
    if lv1 == 1: # Si el nivel es fácil
        n = "4f"
        encabezado2(vidas, pistas_restantes)
        frase(vidas,nom , pistas_restantes,n)
        clear_output()

        # Se genera aleatoriamente la puerta correcta (1–3)
        puerta_correcta = random.randint(1, 3)

        # Mientras el jugador tenga vidas
        while vidas > 0:
            encabezado2(vidas, pistas_restantes)
            print(f"{nom} te encuentras frente a 3 puertas misteriosas. Solo una es segura.\n")
            print("1️⃣ Puerta 1: Pasillo oscuro con un sonido extraño.")
            print("2️⃣ Puerta 2: Luz brillante y cálida.")
            print("3️⃣ Puerta 3: Puerta de metal con símbolos antiguos.\n")

            eleccion = input(f"{nom} elige una puerta (1, 2 o 3) o escribe 'pista' para usar una: ")

            # Si el jugador pide pista
            if eleccion.upper() == "PISTA":
                if pistas_restantes > 0:
                    pistas_restantes -= 1
                    print(f"💡 Pista: La puerta correcta NO es la puerta {random.choice([i for i in [1,2,3] if i != puerta_correcta])}")
                    time.sleep(4)
                    clear_output()
                else:
                    print("No te quedan pistas.")
                    clear_output()
                    continue

            # Si la entrada es inválida
            if eleccion not in ["1", "2", "3","PISTA"]:
                print("Entrada inválida, elige 1, 2 o 3.")
                clear_output()
                continue

            eleccion = int(eleccion)
            # Si elige la puerta correcta → gana
            if eleccion == puerta_correcta:
                print(f"\n🎉¡Correcto! La puerta {eleccion} te lleva al siguiente nivel.\n\nHas escapado exitosamente.\n")
                time.sleep(8)
                clear_output()
                return vidas, pistas_restantes
            else: # Si falla → pierde una vida
                vidas -= 1
                print(f"\n💥MAL! La puerta {eleccion} estaba equivocada.\n\nTe quedan {vidas} vidas❤️.\n")
                time.sleep(4)
                clear_output()
                if vidas == 0:
                    return vidas, pistas_restantes
    else: # Si el nivel es dificl
        n = "4d"

        encabezado2(vidas, pistas_restantes)
        frase(vidas,nom , pistas_restantes,n)
        clear_output()

        # Penalización inicial en modo difícil
        if pistas_restantes > 0:
            pistas_restantes-= 1
        if vidas == 2:
            vidas -= 1
        elif vidas == 3:
            vidas -= 2

        # Mientras tenga vidas
        while vidas > 0:
            encabezado2(vidas, pistas_restantes)
            print(f"{nom} te encuentras frente a 2 puertas misteriosas. Solo una es segura,\n"+"50/50, solo una persona temeraria como tú toma una decisión rápida y fria.\n")
            print("1️⃣ Puerta 1: Zoológico abandonado por los humanos.")
            print("2️⃣ Puerta 2: Un pueblo tranquilo en brasil.")

            puerta_correcta = 1
            pistas = ["Una puerta tiene animales agresivos","Una puerta puerta esta apunto de pasar a la extincion"]
            pista_elegida = random.choice(pistas)

            respuesta_usuario = input(f"\n{nom}, elige una puerta (1 o 2) o escribe 'pista' para usar una: ")

            # Si pide pista
            if respuesta_usuario.upper() == "PISTA":
                if pistas_restantes > 0:
                    pistas_restantes -= 1
                    print(f"💡Pista: {pista_elegida}")
                    time.sleep(4)
                    clear_output()
                else:
                    print("No te quedan pistas.")
                    clear_output()
                    continue

            # Si la entrada es inválida
            if respuesta_usuario not in ["1", "2","PISTA"]:
                print("Entrada inválida, elige 1 o 2.")
                clear_output()
                continue

            respuesta_usuario = int(respuesta_usuario)
            # Si elige la puerta correcta → gana
            if respuesta_usuario == puerta_correcta:
                print(f"\n🎉¡Correcto! La puerta {respuesta_usuario} tenia animales peligrosos, lamentablemnte\nmurieron de hambre.\n\nHas escapado exitosamente.\n")
                time.sleep(8)
                clear_output()
                return vidas, pistas_restantes
            else: # Si falla → pierde vida
                vidas -= 1
                print(f"\n💥MAL! La puerta {respuesta_usuario} te llevo a un pueblo tranquilo, donde en 3 segundos\nva a caer un meteorito, que contamina el aire y mueres por respirar.\n\nTe quedan {vidas} vidas❤️.\n")
                time.sleep(8)
                clear_output()
                if vidas == 0:
                    return vidas, pistas_restantes
```
